# Step Test Simulations

```python
exp = {'04_Step_Data/Step40': 40,
       '04_Step_Data/Step45': 45,
       '04_Step_Data/Step50': 50,
       '04_Step_Data/Step55': 55,
       '04_Step_Data/Step60': 60}

E = exp.keys()
```

```python
from tclab import TCLab, Historian, Plotter, clock, setup
TCLab = setup(connected=False, speedup=10)

import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline

tfinal = 700

for e in E:
    with TCLab() as a:
        h = Historian(a.sources)
        p = Plotter(h,tfinal)
        a.U1 = exp[e]
        for t in clock(tfinal):
            p.update(t)

    df = pd.DataFrame.from_records(h.log, columns=h.columns, index='Time')
    df.to_csv(e)
```
